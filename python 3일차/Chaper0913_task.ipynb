__[문제0]__


오즈 코딩 스쿨 백엔드 8기 과정에 참여한 수강생 중 3명의 수강생이 자율적으로 스터디 그룹을 결성했습니다.

스터디 그룹 내에는 한 주간 배운 내용으로 매주 쪽지 시험을 보는 룰이 있습니다. 점수는 채점자 마음입니다.

쪽지 시험의 점수를 직접 입력받고, 점수의 합을 출력해주세요!


<br>



````python
# 한줄씩 따로 입력 (사용자가 헷갈리지않고 않게 입력가능)

score1 = int(input("철수의 점수를 정수로 입력하시오 : "))
score2 = int(input("영수의 점수를 정수로 입력하시오 : "))
score3 = int(input("백수의 점수를 정수로 입력하시오 : "))

total_score = (score1 + score2 + score3)
print (f" 3명의 총 합은 {total_score} 점")
````

````python
# 한줄에 입력 (입력이 간편하고 빠르나 사용자가 입력형식을 모르면 오류발)

scores = input("점수를 공백으로 구분하여 입력하세요 (예: 80 70 70): ").split()
score1, score2, score3 = map(int, scores)

total_score = score1 + score2 + score3

print (f" 3명의 총 합은 {total_score} 점")
````

__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]

max_num = max(nums)
min_num = min(nums)
average = sum(nums) / len(nums)

print(f"최댓값: {max_num}")
print(f"최솟값: {min_num}")
print(f"평균: {round(average, 2)}") # round를 사용하여 소수점을 깔끔하게 정리할수있다 추후 활용더해보기. 소수점 2개까지
````

__[문제6]__

<br>
다음 딕셔너리를 사용하여 메로나 가격을 출력해주세요

```ice = {'메로나': 1000,```<br>
&emsp;&emsp;&emsp;&emsp;```'폴로포': 1200,```<br>
&emsp;&emsp;&emsp;&emsp;```'빵빠레': 1800,```<br>
&emsp;&emsp;&emsp;&emsp;```'죠스바': 1200,```<br>
&emsp;&emsp;&emsp;&emsp;```'월드콘': 1500}```<br>

<br>
출력 예) 메로나 가격: 1000

````python
ice = {
    '메로나': 1000,
    '폴로포': 1200,
    '빵빠레': 1800,
    '죠스바': 1200,
    '월드콘': 1500
}

print (f"메로나 가격은 {ice['메로나']} 원 입니다")
````

````python
# 아이스크림 딕셔너리
ice = {
    '메로나': 1000,
    '폴로포': 1200,
    '빵빠레': 1800,
    '죠스바': 1200,
    '월드콘': 1500
}

# 사용자로부터 아이스크림 이름 입력받기
ice_name = input("가격을 알고 싶은 아이스크림 이름을 입력하세요: ")

# 입력된 아이스크림이 딕셔너리에 있는지 확인
if ice_name in ice:
    print(f"{ice_name}의 가격은 {ice[ice_name]} 원입니다.")
else:
    print("해당 아이스크림은 없습니다.")
````
