__[과제 1: 초기화 메서드 __init__]__

목표: __init__ 메서드를 구현하여 붕어빵 가게의 초기 재고, 가격, 판매량을 설정합니다.

설명: self.stock, self.prices, self.sales 딕셔너리를 초기화하세요.
- stock
    - "팥붕어빵": 10
    - "슈크림붕어빵": 8
    - "초코붕어빵": 5

- prices
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- sales
    - "팥붕어빵": 0,
    - "슈크림붕어빵": 0,
    - "초코붕어빵": 0

```python
def __init__(self):

        # 초기 재고 설정
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }

        # 붕어빵 가격 설정
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }

        # 붕어빵 판매량 설정
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }
```

__[과제 2: 재고 확인 메서드 check_stock]__

목표: check_stock 함수를 구현하여 현재 붕어빵 재고를 출력합니다.

설명: self.stock 딕셔너리를 순회하며 각 붕어빵의 종류와 재고를 출력하세요.


```python
def check_stock(self):
        print("현재 재고:")
        for bread, count in self.stock.items():
            print(f"{bread}: {count}개")
```

__[과제 3: 주문 처리 메서드 process_order]__

목표: process_order 함수를 구현하여 손님의 주문을 처리하고 재고를 업데이트합니다.

설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
def process_order(self, bread_type, bread_count):
        if bread_type in self.stock:
            if self.stock[bread_type] >= bread_count:
                self.stock[bread_type] -= bread_count
                self.sales[bread_type] += bread_count
                print(f"{bread_type} {bread_count}개를 판매했습니다.")
            else:
                print(f"{bread_type} 재고가 부족합니다. 현재 재고는 {self.stock[bread_type]}개입니다.")
        else:
            print("존재하지 않는 붕어빵 종류입니다.")
```

__[과제 4: 관리자 모드 메서드 admin_mode]__

목표: admin_mode 메서드를 구현하여 관리자 모드에서 재고를 추가할 수 있도록 합니다.

설명: bread_type과 additional_stock을 입력받아 재고를 추가합니다. 잘못된 입력 값이 들어온 경우 적절한 메시지를 출력하세요.

```python
def admin_mode(self):
        while True:
            bread_type = input("재고를 추가할 붕어빵 종류를 입력하세요 (종료하려면 '종료' 입력): ")
            if bread_type == "종료":
                break
            if bread_type in self.stock:
                try:
                    additional_stock = int(input(f"{bread_type}에 추가할 재고 수량을 입력하세요: "))
                    if additional_stock > 0:
                        self.stock[bread_type] += additional_stock
                        print(f"{bread_type}의 재고가 {additional_stock}개 추가되었습니다.")
                    else:
                        print("재고 수량은 0보다 커야 합니다.")
                except ValueError:
                    print("유효한 숫자를 입력해주세요.")
            else:
                print("존재하지 않는 붕어빵 종류입니다.")
```


__[과제 5: 매출 계산 메서드 calculate_total_sales]__

목표: calculate_total_sales 메서드를 구현하여 총 매출을 계산하고 출력합니다.

설명: self.sales와 self.prices를 사용하여 총 매출을 계산하고 출력하세요.

```python
def calculate_total_sales(self):
        total_sales = 0
        total_count = 0
        for bread_type, sold_count in self.sales.items():
            total_sales += sold_count * self.prices[bread_type]
            total_count += sold_count
            print(f"{bread_type}: {sold_count}개 판매, 매출: {sold_count * self.prices[bread_type]}원")
        print(f"총 판매 개수는 {total_count}개이고, 총 매출은 {total_sales}원입니다.")
```

__[과제 6: 메인 함수 main]__

목표: main 함수를 작성하여 프로그램의 주요 로직을 실행합니다.

설명:
- BungeoppangShop 객체를 생성합니다.
- 사용자에게 모드를 선택하도록 요청합니다.
- "주문" 모드에서는 손님의 주문을 처리하고, "관리자" 모드에서는 재고를 추가할 수 있습니다.
- 프로그램이 종료될 때 총 매출을 계산하고 데이터를 저장합니다.


```python
def main():
    shop = BungeoppangShop()

    while True:
        mode = input("모드를 선택하세요 (주문 / 관리자 / 종료): ")

        if mode == "주문":
            bread_type = input("주문할 붕어빵 종류를 입력하세요: ")
            try:
                bread_count = int(input("주문할 개수를 입력하세요: "))
                shop.process_order(bread_type, bread_count)
            except ValueError:
                print("유효한 숫자를 입력해주세요.")

        elif mode == "관리자":
            shop.admin_mode()

        elif mode == "종료":
            print("프로그램을 종료합니다.")
            shop.calculate_total_sales()
            break

        else:
            print("유효한 모드를 입력해주세요. (주문 / 관리자 / 종료)")

if __name__ == "__main__":
    main()
```

__[과제 7: 최종 BungeoppangShop 클래스]__

함수로 리팩토링한 "붕어빵 자동 판매 시스템" 작성 후 실행합니다.

```python
class BungeoppangShop:
    def __init__(self):
        # 초기 재고 설정
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }

        # 붕어빵 가격 설정
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }

        # 초기 판매량 설정
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }

    def check_stock(self):
        print("현재 재고:")
        for bread, count in self.stock.items():
            print(f"{bread}: {count}개")

    def process_order(self, bread_type, bread_count):
        if bread_type in self.stock:
            if self.stock[bread_type] >= bread_count:
                self.stock[bread_type] -= bread_count
                self.sales[bread_type] += bread_count
                print(f"{bread_type} {bread_count}개를 판매했습니다.")
            else:
                print(f"{bread_type} 재고가 부족합니다. 현재 재고는 {self.stock[bread_type]}개입니다.")
        else:
            print("존재하지 않는 붕어빵 종류입니다.")

    def admin_mode(self):
        while True:
            bread_type = input("재고를 추가할 붕어빵 종류를 입력하세요 (종료하려면 '종료' 입력): ")
            if bread_type == "종료":
                break
            if bread_type in self.stock:
                try:
                    additional_stock = int(input(f"{bread_type}에 추가할 재고 수량을 입력하세요: "))
                    if additional_stock > 0:
                        self.stock[bread_type] += additional_stock
                        print(f"{bread_type}의 재고가 {additional_stock}개 추가되었습니다. 현재 재고는 {self.stock[bread_type]}개입니다.")
                    else:
                        print("재고 수량은 0보다 커야 합니다.")
                except ValueError:
                    print("유효한 숫자를 입력해주세요.")
            else:
                print("존재하지 않는 붕어빵 종류입니다.")

    def calculate_total_sales(self):
        total_sales = 0
        total_count = 0
        total_remaining_stock = 0

        for bread_type, sold_count in self.sales.items():
            total_sales += sold_count * self.prices[bread_type]
            total_count += sold_count
            print(f"{bread_type}: {sold_count}개 판매, 매출: {sold_count * self.prices[bread_type]}원, 남은 재고: {self.stock[bread_type]}개")
            total_remaining_stock += self.stock[bread_type]

        print(f"총 판매 개수는 {total_count}개이고, 총 매출은 {total_sales}원입니다.")

def main():
    shop = BungeoppangShop()

    while True:
        mode = input("모드를 선택하세요 (주문 / 관리자 / 종료): ")

        if mode == "주문":
            bread_type = input("주문할 붕어빵 종류를 입력하세요: ")
            try:
                bread_count = int(input("주문할 개수를 입력하세요: "))
                shop.process_order(bread_type, bread_count)
            except ValueError:
                print("유효한 숫자를 입력해주세요.")

        elif mode == "관리자":
            shop.admin_mode()

        elif mode == "종료":
            print("프로그램을 종료합니다.")
            shop.calculate_total_sales()
            break

        else:
            print("유효한 모드를 입력해주세요. (주문 / 관리자 / 종료)")

if __name__ == "__main__":
    main()
```
