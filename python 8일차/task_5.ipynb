[문제]

'Account' 클래스를 작성하세요.

이 클래스는 소유자(owner)와 잔액(balance)을 속성으로 가지며, 잔액은 프라이빗 변수로 선언합니다.

'deposit'와 'withdraw' 두 개의 메소드를 통해 잔액을 조작할 수 있도록 하세요.

이 때, 잔액은 음수가 될 수 없도록 체크해야 합니다. 'Account' 객체를 생성하고,<br>
입금 및 출금 작업을 시뮬레이션한 후, 최종 잔액을 출력하세요.

```python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    # 입금
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"{amount}원이 입금되었습니다. 현재 잔액: {self.__balance}원")
        else:
            print("입금 금액은 0보다 커야 합니다.")

    # 출금
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"{amount}원이 출금되었습니다. 현재 잔액: {self.__balance}원")
        else:
            print("출금 실패: 잔액이 부족합니다.")

    # 잔액 조회
    def get_balance(self):
        return self.__balance

# Account 객체 생성
account = Account("홍길동", 10000)

# 입금 및 출금 작업
account.deposit(5000)    # 입금
account.withdraw(3000)   # 출금
account.withdraw(15000)  # 실패

# 최종 잔액
print(f"최종 잔액: {account.get_balance()}원")
```
