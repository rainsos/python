__[문제 2: 붕어빵 주문 받기]__

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
inventory = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

# 주문을 저장할 딕셔너리
order = {}

while True:
    bread = input("주문할 붕어빵 종류를 입력하세요 (종료: '종료'): ")

    # 종료 입력 시 반복문 종료
    if bread == "종료":
        break

    # 재고에 없는 붕어빵이면 다시 입력받기
    if bread not in inventory:
        print("해당 붕어빵은 없습니다. 다시 입력해주세요.")
        continue

    # 주문 개수 입력
    count = int(input(f"{bread} 몇 개 주문하시겠습니까?: "))

    # 주문 내역에 저장 (이미 주문한 항목이라면 수량 추가)
    if bread in order:
        order[bread] += count
    else:
        order[bread] = count

# 주문 내역 출력
print("\n--- 주문 내역 ---")
for bread, count in order.items():
    print(f"{bread}: {count}개")
```

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
while True:
    bread = input("주문할 붕어빵 종류를 입력하세요 (종료: '종료'): ")

    # 종료 입력 시 반복문 종료
    if bread == "종료":
        break

    # 없는 붕어빵일 경우 안내
    if bread not in inventory:
        print("해당 붕어빵은 없습니다. 다시 입력해주세요.")
        continue

    # 주문 개수 입력
    count = int(input(f"{bread} 몇 개 주문하시겠습니까?: "))
    order[bread] = count

# 주문 처리
print("\n--- 주문 처리 결과 ---")
for bread, count in order.items():
    # 재고가 충분한 경우
    if inventory[bread] >= count:
        inventory[bread] -= count
        print(f"{bread}: {count}개 판매 완료! 남은 재고: {inventory[bread]}개")
    else:
        # 재고가 부족한 경우
        print(f"{bread}: 재고 부족! 현재 재고: {inventory[bread]}개, 주문 수량: {count}개")

# 최종 재고 출력
print("\n--- 최종 재고 ---")
for bread, stock in inventory.items():
    print(f"{bread}: {stock}개")
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
# 초기 재고 설정
stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

# 주문 처리 반복
while True:
    print("\n현재 재고:")
    for bread, count in stock.items():
        print(f"{bread}: {count}개")

    bread_type = input("\n주문할 붕어빵 종류를 입력하세요 (종료: '종료'): ")

    # 종료 조건
    if bread_type == "종료":
        print("\n주문을 종료합니다.")
        break

    # 없는 메뉴일 경우
    if bread_type not in stock:
        print("해당 붕어빵은 판매하지 않는 메뉴입니다.")
        continue

    # 주문 수량 입력
    try:
        bread_count = int(input(f"{bread_type} 몇 개 주문하시겠습니까?: "))
    except ValueError:
        print("숫자만 입력해주세요!")
        continue

    # 0 이하의 숫자는 받지 않도록 처리
    if bread_count <= 0:
        print("주문 수량은 1개 이상이어야 합니다.")
        continue

    # 재고 확인 및 업데이트
    if stock[bread_type] >= bread_count:
        stock[bread_type] -= bread_count
        print(f"{bread_type} {bread_count}개를 판매했습니다.")
    else:
        print(f"죄송합니다. {bread_type} 재고가 부족합니다. 현재 재고: {stock[bread_type]}개")
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
# 관리자 모드 시작
while True:
    print("\n--- 현재 재고 ---")
    for bread, count in stock.items():
        print(f"{bread}: {count}개")

    # 추가할 붕어빵 입력
    bread_type = input("\n재고를 추가할 붕어빵을 입력하세요 (종료: '종료'): ")

    # 종료 조건
    if bread_type == "종료":
        print("\n관리자 모드를 종료합니다.")
        break

    # 없는 메뉴일 경우 처리
    if bread_type not in stock:
        print("해당 붕어빵은 판매하지 않는 메뉴입니다.")
        continue

    # 추가할 수량 입력
    try:
        add_count = int(input(f"{bread_type} 몇 개를 추가하시겠습니까?: "))
    except ValueError:
        print("숫자만 입력해주세요!")
        continue

    # 0 이하 수량은 처리하지 않음
    if add_count <= 0:
        print("추가 수량은 1개 이상이어야 합니다.")
        continue

    # 재고 업데이트
    stock[bread_type] += add_count
    print(f"{bread_type} 재고가 {add_count}개 추가되었습니다. 현재 재고: {stock[bread_type]}개")
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
# 붕어빵 가격 설정
prices = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}

# 판매된 붕어빵 개수 초기화
sales = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}

# 판매 데이터 입력 받기
while True:
    print("\n--- 현재 판매 현황 ---")
    for bread, count in sales.items():
        print(f"{bread}: {count}개 판매됨")

    bread_type = input("\n판매된 붕어빵 종류를 입력하세요 (종료: '종료'): ")

    # 종료 조건
    if bread_type == "종료":
        print("\n매출 확인을 종료합니다.")
        break

    # 없는 메뉴일 경우 처리
    if bread_type not in prices:
        print("해당 붕어빵은 판매하지 않는 메뉴입니다.")
        continue

    # 판매 개수 입력
    try:
        sell_count = int(input(f"{bread_type} 몇 개를 판매하셨습니까?: "))
    except ValueError:
        print("숫자만 입력해주세요!")
        continue

    # 0 이하 수량은 처리하지 않음
    if sell_count <= 0:
        print("판매 수량은 1개 이상이어야 합니다.")
        continue

    # 판매 개수 업데이트
    sales[bread_type] += sell_count
    print(f"{bread_type} {sell_count}개 판매가 기록되었습니다.")

# 매출 계산
total_sales = 0
print("\n--- 매출 확인 ---")
for bread, count in sales.items():
    sale_amount = prices[bread] * count
    total_sales += sale_amount
    print(f"{bread}: {count}개 판매, 매출: {sale_amount}원")

print(f"\n총 매출: {total_sales}원")
```
