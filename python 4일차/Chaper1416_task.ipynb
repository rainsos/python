__[문제1]__

주어진 정수 x와 자연수 n을 이용해, x부터 시작해 x씩 증가하는 숫자를 n개 지니는 리스트를 출력해주세요   
<br>

x = 2, n =5

```python
# 시작 숫자(x)과 갯수(n)

x = 2
n = 5

# 결과를 저장할 빈 리스트
결과 = []

# 반복문으로 리스트에 숫자를 추가

for i in range(1,+n+1):
    결과.append(x * i)

# 결과 출력
print(결과)
```

__[문제3]__

숫자를 건넬 때 일부 자릿수를 영단어로 바꾼 카드를 건네주면 프로도는 원래 숫자를 찾는 게임입니다.

다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

1478 → "one4seveneight"  
234567 → "23four5six7"  
10203 → "1zerotwozero3"    

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나,   
혹은 바뀌지 않고 그대로인 문자열 s가 주어졌을때
s가 의미하는 원래 숫자를 출력하도록 코드를 작성해주세요

s = "77three4one

---
참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

숫자	영단어  
0	zero  
1	one  
2	two  
3	three  
4	four  
5	five  
6	six  
7	seven  
8	eight  
9	nine  

```python
num_dict = {
    "zero": "0",
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4",
    "five": "5",
    "six": "6",
    "seven": "7",
    "eight": "8",
    "nine": "9"
}
s = "77three4one"
for key in num_dict.keys(): # 햇갈리지않기 이것은 dict에 있는 모든 key값을 가져온다는것
    s = s.replace(key, num_dict[key]) # 이것은 딕셔너리에있는 key값에 대응하는 value를 가지고오는것

print(s)
```

__[문제4]__

행렬의 덧셈은 행과 열의 크기가 같은 두 행렬의 같은 행, 같은 열의 값을 서로 더한 결과가 됩니다.

 행렬 arr1과 arr2가 주어졌을 때, 행렬 덧셈의 결과를 출력하는 코드를 작성해주세요(result와 같은 결과가 출력되도록 작성)

arr1 = [[1,2],[2,3]]

arr2 = [[3,4],[5,6]]

result = [[4,6],[7,9]]

```python
arr1 = [[1,2],[2,3]]
arr2 = [[3,4],[5,6]]

result = []
for i in range(len(arr1)):
    row = []
    for j in range(len(arr1[0])):
        row.append(arr1[i][j] + arr2[i][j])
    result.append(row)

print(result)
```

```python
arr1 = [[1,2],[2,3]]
arr2 = [[3,4],[5,6]]

result = [[c1 + c2 for c1, c2 in zip(r1, r2)] for r1, r2 in zip(arr1, arr2)]

print(result)
```

__[문제7]__

0부터 9까지의 숫자 중 일부가 들어있는 정수 배열 numbers가 주어집니다.   
numbers에서 찾을 수 없는 0부터 9까지의 숫자를 모두 찾아 더한 수를 출력해주세요

numbers = [5,8,1,0,6,9]

```python
# 기술블로그 참조 https://blog.naver.com/jangpanjangsu/223792638003

numbers = [5,8,1,0,6,9]

full_set = set(range(10))  # 0부터 9까지의 숫자를 포함하는 집합을 생성
numbers_set = set(numbers)  # 주어진 배열을 집합으로 변환
missing_numbers = full_set - numbers_set  # 두 집합의 차집합을 구함
result = sum(missing_numbers)  # 차집합의 모든 원소를 더함

print(result)
```
