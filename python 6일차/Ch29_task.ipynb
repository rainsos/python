주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
# 리턴 (둘중에 어떤것이 조기리턴이라고 불릴수있는건지?)

def is_even(number: int) -> bool:
    if number % 2 == 0:
        return True
    return False

print(is_even(4))
print(is_even(7))
```

```python
# 리턴 (둘중에 어떤것이 조기리턴이라고 불릴수있는건지?)

def is_even(number: int) -> bool:
    return number % 2 == 0

print(is_even(4))
print(is_even(7))
```
