사용자로부터 이름과 나이를 입력 받아, "안녕하세요, [이름]님! 당신은 [나이]
살입니다."라고 출력하는 함수 greet_user를 작성하세요.

```python
def greet_user(name, age):
    print(f"안녕하세요, {name}님! 당신은 {age}살입니다.")

greet_user("이상인", 39)
```

```python
def greet_user():

    name = input("이름을 입력하세요: ")
    age = input("나이를 입력하세요: ")

    if age.isdigit():
        age = int(age)  # str -> int

        if age <= 0:
            print("나이는 1살 이상으로 입력해 주세요!")
        else:
            print(f"안녕하세요, {name}님! 당신은 {age}살입니다.")
    else:
        print("나이는 숫자(양수)로 입력해 주세요!")

greet_user()
```
