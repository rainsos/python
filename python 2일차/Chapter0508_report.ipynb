<br>

phone = "010-1234-5678"

출력 예시 : "01012345678"

__Tips__
- split() 함수는 문자열을 특정 구분자를 기준으로 나누어 리스트로 반환합니다.

```python
# 전화번호.split('-')는 전화번호를 하이픈('-')을 기준으로 나누어 리스트를 반환하고, ''.join()은 이 리스트의 요소들을 빈 문자열('')로 연결하여 하나의 문자열로 만든다

phone = "010-1234-5678"
print((phone.split('-'))) # 값이 떨어져서 나옴

phone = "010-1234-5678"
print(''.join(phone.split('-'))) # join() 함수를 이용해 다시 합친다
```

__[문제 5: 레벨업]__

```python
# 각 직업 별 경험치
xp = {"wizard": 120, "warrior": 150, "rogue": 110, "healer": 90}

# 출력
print(xp["wizard"] >= 100)
print(xp["warrior"] >= 100)
print(xp["rogue"] >= 100)
print(xp["healer"] >= 100)
```
