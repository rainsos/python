__[문제1]__

어떤 공원의 크기는 가로 50m, 세로 30m입니다.

공원 내부에 5개의 조형물이 있으며, 각 조형물의 면적은 10m²입니다.

이 공원의 실제 사용 면적을 구하세요. (cm로 구해주세요)




```python
((50*30)-(5*10))*10000
```

```python
width = 50  # 공원의 가로 길이 (m)
height = 30  # 공원의 세로 길이 (m)
structures = 5  # 조형물 개수
structure_area = 10  # 조형물 하나의 면적 (m²)

area = (width * height - structures * structure_area) * 10000 # cm²로 전환
print (f"{area} cm²")
```

__[문제2]__

2권씩 묶여서 팔리는 책의 가격은 12,000원입니다.

나는 3묶음을 살 겁니다.

근데 최종 가격을 보니 10%의 세금이 붙어있습니다.

최종 가격을 구해주세요.

```python
12000*3*1.1
```

```python
price = 12_000  # 2권 묶음의 가격
bundles = 3  # 구매할 묶음 수
tax = 0.1  # 세율 (10%)

final_price = price * bundles * (1 + tax)

print (str(final_price)+"원")
```
